```python
import random
import datetime

# instance parameters
n_S = random.randint(1, 20)
n_C = random.randint(1, 50)
n_L = random.randint(1, 10)
n_K = random.randint(1, 50)
n_D = random.randint(1, 10)
B = random.randint(1, 10000)

# create cars data
car_ids = list(range(1, n_C + 1))
car_levels = [random.randint(1, n_L) for _ in range(n_C)]
car_init_stations = [random.randint(1, n_S) for _ in range(n_C)]
with open('data\instance00.txt', 'w') as f:
    f.write('n_C,n_L,n_S,n_K,n_D,B\n')
    f.write(f'{n_C},{n_L},{n_S},{n_K},{n_D},{B}\n')
    f.write('==========\n')
    f.write('Car ID,Level,Initial station\n')
    for i in range(n_C):
        f.write(f'{car_ids[i]},{car_levels[i]},{car_init_stations[i]}\n')
    f.write('==========\n')
    f.write('Car level,Hour rate\n')
    for i in range(1, n_L + 1):
        f.write(f'{i},{random.randint(50, 500)}\n')
    f.write('==========\n')

# create orders data
order_ids = list(range(1, n_K + 1))
order_levels = [random.randint(1, n_L) for _ in range(n_K)]
order_pickup_stations = [random.randint(1, n_S) for _ in range(n_K)]
order_return_stations = [random.randint(1, n_S) for _ in range(n_K)]
order_pickup_times = [datetime.datetime.now() + datetime.timedelta(days=random.randint(1, 365)) for _ in range(n_K)]
order_return_times = [order_pickup_times[i] + datetime.timedelta(hours=random.randint(1, 72)) for i in range(n_K)]
with open('data\instance00.txt', 'a') as f:
    f.write('Order ID,Level,Pick-up station,Return station,Pick-up time,Return time\n')
    for i in range(n_K):
        f.write(f'{order_ids[i]},{order_levels[i]},{order_pickup_stations[i]},{order_return_stations[i]},')
        f.write(f'{order_pickup_times[i].strftime("%Y/%m/%d %H:%M")},{order_return_times[i].strftime("%Y/%m/%d %H:%M")}\n')
    f.write('==========\n')

# create distances data
distances = [[0] * n_S for _ in range(n_S)]
for i in range(n_S):
    for j in range(n_S):
        if i != j:
            distances[i][j] = random.randint(1, 100)*30
with open('data\instance00.txt', 'a') as f:
    f.write('==========\n')
    f.write('From,To,Distance\n')
    for i in range(n_S):
        for j in range(n_S):
            f.write(f'{i+1},{j+1},{distances[i][j]}\n')
    f.write('==========\n')

f.close()
```
