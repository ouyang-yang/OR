```python
from gurobipy import *

# Input
I = 5
T = 10
destinations = [(2, 10, 0), (8, 20, 0), (9, 21, 1), (6, 14, 1)]

# Create model
m = Model()

# Create variables
stay = {}
pass_through = {}
for i in range(1, I):
    stay[i] = m.addVar(vtype=GRB.BINARY, name="stay_{}".format(i))
    pass_through[i] = m.addVar(vtype=GRB.BINARY, name="pass_through_{}".format(i))

# Create objective
obj = LinExpr()
for i in range(1, I):
    obj += (stay[i] + pass_through[i]) * destinations[i-1][1] # satisfaction
    obj += (stay[i] + pass_through[i]) * destinations[i-1][0] # time spent
m.setObjective(obj, GRB.MAXIMIZE)

# Create constraints
m.addConstr(quicksum(stay[i]*destinations[i-1][0] + pass_through[i]*destinations[i-1][0] for i in range(1, I)) <= T, "time_constraint")

for i in range(2, I):
    m.addConstr(stay[i] + pass_through[i] <= quicksum(stay[j] for j in range(1, i) if destinations[j-1][2] == i-2) + quicksum(pass_through[j] for j in range(1, i) if destinations[j-1][2] == i-2), "connectivity_constraint_1_{}".format(i))
    m.addConstr(stay[i] + pass_through[i] <= quicksum(stay[j] for j in range(i, I) if destinations[j-1][2] == i-1) + quicksum(pass_through[j] for j in range(i, I) if destinations[j-1][2] == i-1), "connectivity_constraint_2_{}".format(i))

# Solve model
m.optimize()

# Print solution
if m.status == GRB.OPTIMAL:
    print("Optimal solution found: {}".format(int(m.objVal)))
else:
    print("No solution found.")
```
