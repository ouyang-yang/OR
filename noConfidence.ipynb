```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import csv
```

```python
def load_info(path):
    info = {}
    with open(r'C:\Users\mayda\Python practice\OR\LP_results_new.csv') as f:
        myCsv = csv.reader(f)
        for row in myCsv:
            key = row[0]
            values = np.array(row[1:], dtype=float)
            info[key] = values

    B = info.get('B', [0])[0]
    L = info.get('L', [0])[0]
    w = info.get('w', [0])[0]
    obj_value = info.get('obj. value', [0])[0]
    var_angle = info.get('var\\angle', [])
    X = np.flip(info.get('X', []))
    Xe = np.flip(info.get('Xe', []))
    E = np.flip(info.get('E', []))
    XBev = np.flip(info.get('XBev', []))
    XeBev = np.flip(info.get('XeBev', []))
    EBev = np.flip(info.get('EBev', []))
    XMot = np.flip(info.get('XMot', []))
    XeMot = np.flip(info.get('XeMot', []))
    EMot = np.flip(info.get('EMot', []))

    return (
        B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot
    )
```

```python
def draw_parking_lot(B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot):
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))

    axs[0, 0].set_title('Parking Lot')
    axs[0, 0].set_xlabel('var\\angle')
    axs[0, 0].set_ylabel('X')
    axs[0, 0].plot(var_angle, X, marker='o')
    axs[0, 0].plot(var_angle, Xe, marker='o')
    axs[0, 0].legend(['X', 'Xe'])

    axs[0, 1].set_title('Energy')
    axs[0, 1].set_xlabel('var\\angle')
    axs[0, 1].set_ylabel('E')
    axs[0, 1].plot(var_angle, E, marker='o')
    axs[0, 1].plot(var_angle, XBev, marker='o')
    axs[0, 1].plot(var_angle, XeBev, marker='o')
    axs[0, 1].plot(var_angle, EBev, marker='o')
    axs[0, 1].legend(['E', 'XBev', 'XeBev', 'EBev'])

    axs[1, 0].set_title('Motion')
    axs[1, 0].set_xlabel('var\\angle')
    axs[1, 0].set_ylabel('Motion')
    axs[1, 0].plot(var_angle, XMot, marker='o')
    axs[1, 0].plot(var_angle, XeMot, marker='o')
    axs[1, 0].plot(var_angle, EMot, marker='o')
    axs[1, 0].legend(['XMot', 'XeMot', 'EMot'])

    axs[1, 1].axis('off')
    axs[1, 1].text(0.1, 0.8, f'B: {B}', fontsize=12)
    axs[1, 1].text(0.1, 0.7, f'L: {L}', fontsize=12)
    axs[1, 1].text(0.1, 0.6, f'w: {w}', fontsize=12)
    axs[1, 1].text(0.1, 0.5, f'Objective Value: {obj_value}', fontsize=12)

    plt.tight_layout()
    plt.show()
```

```python
def Draw(path):
    B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot = load_info(path)
    draw_parking_lot(B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot)
```

```python
Draw(r'C:\Users\mayda\Python practice\OR\LP_results_new.csv')
```
