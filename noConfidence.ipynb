```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import csv
```

```python
def load_info(path):
    info = {}
    with open(r'C:\Users\mayda\Python practice\OR\LP_results_new.csv') as f:
        myCsv = csv.reader(f)
        for row in myCsv:
            key = row[0]
            values = np.array(row[1:], dtype=float)
            info[key] = values

    B = info.get('B', [0])[0]
    L = info.get('L', [0])[0]
    w = info.get('w', [0])[0]
    obj_value = info.get('obj. value', [0])[0]
    var_angle = info.get('var\\angle', [])
    X = np.flip(info.get('X', []))
    Xe = np.flip(info.get('Xe', []))
    E = np.flip(info.get('E', []))
    XBev = np.flip(info.get('XBev', []))
    XeBev = np.flip(info.get('XeBev', []))
    EBev = np.flip(info.get('EBev', []))
    XMot = np.flip(info.get('XMot', []))
    XeMot = np.flip(info.get('XeMot', []))
    EMot = np.flip(info.get('EMot', []))

    return (
        B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot
    )
```

```python
def draw_parking_lot(B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot):
    plt.figure(figsize=(10, 6))

    # Parking Lot Layout
    plt.subplot(2, 2, 1)
    plt.title('Parking Lot')
    plt.xlabel('Angle')
    plt.ylabel('X')
    plt.plot(var_angle, X, marker='o')
    plt.plot(var_angle, Xe, marker='o')
    plt.legend(['X', 'Xe'])

    # Energy Consumption
    plt.subplot(2, 2, 2)
    plt.title('Energy')
    plt.xlabel('Angle')
    plt.ylabel('Energy')
    plt.plot(var_angle, E, marker='o')
    plt.plot(var_angle, XBev, marker='o')
    plt.plot(var_angle, XeBev, marker='o')
    plt.plot(var_angle, EBev, marker='o')
    plt.legend(['E', 'XBev', 'XeBev', 'EBev'])

    # Motion
    plt.subplot(2, 2, 3)
    plt.title('Motion')
    plt.xlabel('Angle')
    plt.ylabel('Motion')
    plt.plot(var_angle, XMot, marker='o')
    plt.plot(var_angle, XeMot, marker='o')
    plt.plot(var_angle, EMot, marker='o')
    plt.legend(['XMot', 'XeMot', 'EMot'])

    # Additional Information
    plt.subplot(2, 2, 4)
    plt.axis('off')
    plt.text(0.1, 0.8, f'B: {B}', fontsize=12)
    plt.text(0.1, 0.7, f'L: {L}', fontsize=12)
    plt.text(0.1, 0.6, f'w: {w}', fontsize=12)
    plt.text(0.1, 0.5, f'Objective Value: {obj_value}', fontsize=12)

    plt.tight_layout()
    plt.show()
```

```python
def Draw(path):
    B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot = load_info(path)
    draw_parking_lot(B, L, w, obj_value, var_angle, X, Xe, E, XBev, XeBev, EBev, XMot, XeMot, EMot)
```

```python
Draw(r'C:\Users\mayda\Python practice\OR\LP_results_new.csv')
```
